```python
from collections import defaultdict
import time
```

```python
class Graph:
    def __init__(self):
        self.graph = defaultdict(list)
        self.heouristic = defaultdict(list)

    def addEdge(self, u, v):
        self.graph[u].append(v)

    def addHeuristic(self, u, v):
        self.heouristic[u].append(v)

    def printPath(self, path):
        distance = 0
        for i in path:
            distance += self.heouristic[i][0]
            print(i, end="->")
        print("END")
        print("Jarak: ", distance)

    def greedy(self, start, end):

        flag = 0
        path = []
        startTime = time.time()
        path.append(start)
        while(True):
            if start == end:
                flag = 1
                break
            min = 999999
            for i in self.graph[start]:
                if self.heouristic[i][0] < min:
                    min = self.heouristic[i][0]
                    start = i
            path.append(start)

        if flag == 1:
            self.printPath(path)
        else:
            print("Path not found")

        endTime = time.time()
        print("Time: ", endTime - startTime)
```
